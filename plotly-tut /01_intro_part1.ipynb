```python
import plotly.express as px

data = px.data.iris()
data.head()
```

```python
data.describe()
```

```python
data.describe(include='object')
```

```python
px.scatter(data_frame=data, x='sepal_length', y='sepal_width', width=500, height=400)
```

```python
px.scatter(data_frame=data,
           x='sepal_length',
           y='sepal_width',
           color='species',
           marginal_x='violin',
           marginal_y='box',
           symbol='species',
           title='Iris Dataset',
           trendline='ols'
)
```

```python
df = px.data.tips()
df.head()
```

```python
df.describe()
```

```python
df.describe(include=['object'])
```

```python
df.head(3)
```

```python
px.scatter(data_frame=df,
           x='total_bill',
           y='tip',
           facet_col='day',
           category_orders={'day':['Thur', 'Fri', 'Sat', 'Sun']},
           trendline='ols',
           color='smoker')
```

```python
px.scatter(data_frame=df,
           x='total_bill',
           y='tip',
           facet_row='time',
           category_orders={'day': ['Thur', 'Fri', 'Sat', 'Sun']},
           trendline='ols',
           color='smoker')
```

```python
data = px.data.iris()
data.head(2)
```

```python
px.scatter_matrix(data, dimensions=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'],
                 color='species',
                 title='Scatter Matrix')
```

```python
px.parallel_coordinates(data, color='species_id')
```

```python
data = px.data.tips()
data.head()
```

```python
px.parallel_categories(data, color='size')
```

```python
df = px.data.gapminder()
df.head()
```

```python
df.describe()
```

```python
df.describe(include='object')
```

```python
df.head(1)
```

```python
px.scatter(df.query('year == 2007'), x='gdpPercap', y='lifeExp', size='pop', color='continent',log_x=True,
           hover_name='country',size_max=60)
```

```python
px.scatter(df, x='gdpPercap', y='lifeExp', size='pop', color='continent', log_x=True,
           hover_name='country', size_max=60, animation_frame='year', range_y=[25,90])
```

```python
px.scatter(df, x='gdpPercap', y='lifeExp', size='pop', color='continent', log_x=True,
           hover_name='country', size_max=60, animation_frame='year', range_y=[25,90], facet_col='continent', animation_group='country')
```

```python
df.head(1)
```

```python
px.line(df.query("country == 'Poland'"), x='year', y='pop')
```

```python
px.line(df.query("continent == 'Europe'"), x='year', y='pop', color='country')
```
