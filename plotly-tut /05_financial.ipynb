```python
def fetch_financial_data(company='AMZN'):
  """
  This function fetch stock market quotations.
  """
  import pandas_datareader.data as web
  return web.DataReader(name=company, data_source='stooq')

df = fetch_financial_data()
df = df.reset_index()
df.head(2)
```

```python
df = df[df.Date > '2025-01-01']
df.describe()
```

```python
import plotly.graph_objects as go
```

```python
fig = go.Figure(
    go.Ohlc(
        x = df.Date,
        open = df.Open,
        high = df.High,
        low = df.Low,
        close = df.Close
    )
)
fig.show()
```

```python
fig = go.Figure(
    go.Candlestick(
        x = df.Date,
        open = df.Open,
        high = df.High,
        low = df.Low,
        close = df.Close
    )
)
fig.show()
```

### Waterfall

```python
go.Waterfall()
```

```python
fig = go.Figure(go.Waterfall(
    x=['Sprzedaż Stacjonarna', 'Sprzedaż Online', 'Przychody', 'Koszty', 'Podatek', 'Zysk Netto'],
    y=[60, 80, 0, -50, -30, 0],
    measure=['relative', 'relative', 'total', 'relative', 'relative', 'total'],
    text =  ['+60', '+80', '', '-50', '-30', ''],
    textposition='outside',
    name='Rachunek zysków i strat 2025'
    )
)
fig.update_layout(
    title='Rachunek zysków i strat',
    showlegend=True
)

fig.show()
```

### Wykres lejkowy

```python
fig = go.Figure(go.Funnel(
    x = [3500, 500, 100 ],
    y = ['Liczba wyświetleń strony', 'Liczba kliknięć', 'Liczba nowych rejestracji'],
    textinfo = 'value+percent initial'
))

fig.show()
```

```python
fig = go.Figure()

fig.add_trace(go.Funnel(
    x = [1000, 500, 50],
    y = ['Liczba wyświetleń strony', 'Liczba kliknięc', 'Liczba nowych rejestracji'],
    textinfo='value+percent initial',
    name='Reklama 1'
))
fig.add_trace(go.Funnel(
    x = [1000, 600, 40],
    y = ['Liczba wyświetleń strony', 'Liczba kliknięc', 'Liczba nowych rejestracji'],
    textinfo='value+percent initial',
    name='Reklama 2'
))
fig.add_trace(go.Funnel(
    x = [2000, 900, 160],
    y = ['Liczba wyświetleń strony', 'Liczba kliknięc', 'Liczba nowych rejestracji'],
    textinfo='value+percent initial',
    name='Reklama 3'
))
fig.show()
```

```python
fig = go.Figure(go.Funnelarea(
    values=[2000, 500, 100],
    text=['Liczba wyświetleń strony', 'Liczba kliknięć', 'Liczba nowych rejestracji']
))
fig.show()
```
