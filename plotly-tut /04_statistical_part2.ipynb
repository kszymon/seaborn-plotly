```python
import plotly.express as px
from seaborn import load_dataset

df = load_dataset('titanic')
df.head(2)
```

```python
px.scatter(df, x='age', y='fare', color='survived', facet_col='sex')
```

```python
px.scatter(df, x='age', y='fare', color='survived', facet_row='sex')
```

```python
px.scatter(df, x='age', y='fare', color='survived', facet_col='sex', facet_row='pclass',
           category_orders={'pclass':[3, 2, 1], 'sex': ['female', 'male']})
```

```python
px.violin(df, x='survived', y='age', color='survived', points='all', box=True)
```
