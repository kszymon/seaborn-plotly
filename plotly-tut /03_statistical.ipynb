```python
from seaborn import load_dataset
import plotly.express as px
import pandas as pd


df = load_dataset('diamonds')
df.head()
```

```python
df.carat = pd.cut(df.carat, bins=10, labels=list('abcedfghij'))
df.head()
```

### Histogram

```python
df = load_dataset('mpg')
df.head()
```

```python
df.describe()
```

```python
df.origin.value_counts()
```

```python
import plotly.express as px

px.histogram(df, x='mpg', nbins=50, title='Rozkad zmiennej mpg')
```

```python
px.histogram(df, x='mpg', nbins=50, title='Rozkad zmiennej mpg', histnorm='probability density')
```

```python
px.histogram(df, x='mpg', nbins=50, title='Rozkad zmiennej mpg',
            facet_row='origin', color='origin')
```

```python
dff = df[(df.origin == 'usa') | (df.origin == 'europe')]
px.histogram(dff, x='mpg', color='origin', marginal='violin', title='Rozkad zmiennej mpg (usa vs. europe)', opacity=0.7)
```

```python
px.histogram(df, x='origin')
```

```python
import plotly.figure_factory as ff

fig = ff.create_distplot([df.mpg], group_labels=['mpg'])
fig.show()
```

```python
fig = ff.create_distplot([df.mpg, df.acceleration], group_labels=['mpg', 'Acceleration'])
fig.show()
```

```python
import numpy as np

x1 = np.random.randn(1000)
x2 = np.random.randn(1000) + 2

ff.create_distplot([x1, x2], ['Rozkad Normalny ~ N(0, 1)', 'Rozkad Normalny ~ N(2, 1)'], curve_type='normal',
                  bin_size=0.2, colors=['slategray', 'magenta'])
```

### Histogram 2D

```python
df = load_dataset('mpg')
df.head()
```

```python
import plotly.graph_objects as go

fig = go.Figure(go.Histogram2d(
    x = df.mpg,
    y = df.acceleration,
    nbinsx = 30,
    nbinsy =30
))

fig.add_trace(go.Scatter(
    x = df.mpg,
    y = df.acceleration,
    mode = 'markers',
    marker = {
        'symbol': 'x',
        'opacity': 0.7,
        'color': 'white',
        'size': 8,
        'line': {
            'width': 1
        }
    }
))

fig.show()
```
